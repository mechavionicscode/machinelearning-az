### **2da Pregunta:** ¿Cuál fue la facturación por categoria em 2023?

**Observación:** Para la facturación, consideraremos aquí el producto entre el precio unitario y la cantidad, sumados al costo de envío de cada producto.


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras horizontales que muestre la facturación por categoría de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categorías de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de envío total de la compra.

Deseamos crear una visualización que muestre la facturación de cada categoría, donde la facturación se calcula multiplicando el precio unitario por la cantidad y sumando el costo de envío.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# calcular la facturacion para cada compra
df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']
````

````python
#Agrupar por categoria y sumar la facturacion
facturacion_categoria = df.groupby('categoria')['facturacion'].sum().reset_index()
````

````python
#Ordenar los datos por facturacion de mayor a menor
facturacion_categoria = facturacion_categoria.sort_values(by='facturacion', ascending=False)
````

````python
#Crear el grafico de barras horizontales
plt.figure(figsize=(10,6))
sns.barplot(x='facturacion', y='categoria', data=facturacion_categoria,palette='viridis',orient='h')

#Añadir etiquetas y titulo
plt.title('Facturacion por Categoria de Producto', fontsize=16)
plt.xlabel('Facturacion (en MXN$)', fontsize=12)
plt.ylabel('Categoria', fontsize=12)

#Mostrar grafico
plt.show()
````

> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de líneas que muestre las ventas totales mensuales de Zoop.

En "df", tenemos las siguientes columnas:
* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturación: calculada como el producto del precio unitario y la cantidad, sumado al costo de envío de cada venta.

Deseamos crear una visualización que muestre las ventas por mes. Primero, agrupa los datos por mes y luego crea una nueva columna llamada "mes" que contenga el nombre de cada mes, traduciendo los meses del índice mediante el siguiente diccionario. Utiliza la columna "fecha" para agrupar los datos y construir el gráfico. La columna "mes" solo debe usarse para cambiar las etiquetas del eje x.

```
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'

```


Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
#1. crear una nueva columna 'mes' que extrae el mes de la colmna 'fecha
df['mes']=df['fecha'].dt.month_name()
````

````python
#2. Agrupar los datos por mes y calcular la facturacion total mensual
ventas_mensuales = df.groupby(df['fecha'].dt.to_period('M'))['facturacion'].sum().reset_index()
````

````python
#3. Convertir el periodo de fecha a tipo datetime para mantener formato de fecha
ventas_mensuales['fecha']=ventas_mensuales['fecha'].dt.to_timestamp()
````

````python
#4. Crear una columna de nombre del mes en español usando el diccionario proporcionado
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}
````

````python
ventas_mensuales ['mes'] = ventas_mensuales['fecha'].dt.month_name().map(meses)
````

````python
#graficos de lineas son buenos para datos temporales continuidad y secuencia de datos
#5. Crear el grafico de lineas de ventas mensuales
plt.figure(figsize=(10,6))
sns.lineplot(x='mes',y='facturacion', data=ventas_mensuales,marker='o',sort=False)

#6. Añadir etiquetas y titulo
plt.title('Ventas Totales Mensuales de Soop', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Facturacion Total', fontsize=12)

#7. Mostrar el grafico
plt.show()
````

Cuando tenemos una cantidad mayor de períodos en nuestros datos, vale la pena recurrir a gráficos de líneas, que son ideales para series temporales. Dado que nuestros datos son mensuales, utilizar un gráfico de columnas, por ejemplo, reduciría la percepción de la secuencia de los datos. Las líneas ofrecen una mayor sensación de continuidad.

La elección entre un histograma de columnas (también conocido como histograma de barras) y un gráfico de densidad generalmente depende del tipo de datos que se analizan y del objetivo de la visualización.

¿Por qué un histograma de columnas?

El histograma de columnas se utiliza ampliamente para visualizar la distribución de frecuencias o conteos en un conjunto de datos. Este tipo de gráfico es particularmente eficaz para mostrar la distribución de variables discretas o categorías. En el caso de las evaluaciones del público, que son valores enteros de 0 a 10, un histograma de columnas permite visualizar claramente cuántas evaluaciones se recibieron para cada punto específico. Esto facilita la identificación de tendencias, como la concentración de evaluaciones en ciertas puntuaciones, y ayuda a comprender la distribución general de las evaluaciones.

Diferencia entre un Histograma de Columnas y un Gráfico de Densidad

* Histograma de Columnas: Utiliza barras verticales para representar la frecuencia de cada categoría o intervalo de valores. La altura de cada barra indica el conteo o la proporción de observaciones en esa categoría.
* Gráfico de Densidad: Utiliza una línea para trazar la distribución continua de los datos. Este tipo de gráfico es más adecuado para variables continuas, donde muestra la probabilidad de valores dentro de un continuo, proporcionando una idea de la densidad de los datos.

¿Cuándo elegir cada uno?

* Histograma de Columnas: Es ideal para datos categóricos o discretos, cuando se desea visualizar la frecuencia o el conteo por categoría. Es útil cuando las categorías son pocas y están claramente definidas, como las calificaciones de 0 a 10.
* Gráfico de Densidad: Es más adecuado para variables continuas, cuando se desea entender la distribución y densidad de los datos a lo largo de un intervalo. Es útil para visualizar la forma de la distribución de los datos, como una distribución normal, binomial, etc.

Ejemplos de Utilización en E-commerce

* Histograma de Columnas: Imagina que deseas analizar las evaluaciones de los clientes para un producto específico en tu e-commerce. Las evaluaciones van de 0 a 5. Un histograma de columnas puede mostrar cuántos clientes dieron cada puntuación, permitiendo identificar rápidamente la satisfacción general de los clientes con el producto.

* Gráfico de Densidad: Si estás analizando el tiempo que los clientes pasan en tu sitio de e-commerce antes de realizar una compra, los datos serían más continuos (por ejemplo, minutos). Un gráfico de densidad podría mostrar la distribución del tiempo de navegación, ayudando a identificar si la mayoría de los clientes compra rápidamente o pasa mucho tiempo navegando antes de decidir.

````python
def box_plot_edad_sexo():
  # configuracin del gráfico
  plt.figure(figsize=(10,6), facecolor=GRIS_1)
  ax =plt.axes()
  ax.set_facecolor(GRIS_1)

  # Graficar el boxplot de la distribucion de la edad por sexo biológico
  sns.boxplot(data=df, x='sexo_biologico', y='edad', palette=[AZUL_1, AQUA_1], ax=ax)

  # Añadir etiquetas y título
  plt.title('Distribucion de la Edad por Sexo Biológico de los Clientes', fontsize=18, color=GRIS_5)
  plt.xlabel('')
  plt.ylabel('Edad', fontsize=12,color=AZUL_1)
  plt.xticks(fontsize=12, color=AZUL_1)
  plt.yticks(fontsize=12, color=AZUL_1)
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.spines['left'].set_visible(False)
  ax.spines['bottom'].set_visible(False)
  ax.set_axisbelow(True)
  ax.yaxis.grid(color='gray',linestyle='--', linewidth=0.5)

  # Añadir caja de anotación
  texto_anotacion1 = (
      f"Mínimo = {df.groupby('sexo_biologico')['edad'].min()[1]}\n"
      f"1º quartil = {df.groupby('sexo_biologico')['edad'].quantile(.25)[1]}\n"
      f"2º quartil = {df.groupby('sexo_biologico')['edad'].median()[1]}\n"
      f"3º quartil = {df.groupby('sexo_biologico')['edad'].quantile(.75)[1]}\n"
      f"Máximo = {df.groupby('sexo_biologico')['edad'].max()[1]}"
  )

  texto_anotacion2 = (
      f"Mínimo = {df.groupby('sexo_biologico')['edad'].min()[0]}\n"
      f"1º quartil = {df.groupby('sexo_biologico')['edad'].quantile(.25)[0]}\n"
      f"2º quartil = {df.groupby('sexo_biologico')['edad'].median()[0]}\n"
      f"3º quartil = {df.groupby('sexo_biologico')['edad'].quantile(.75)[0]}\n"
      f"Máximo = {df.groupby('sexo_biologico')['edad'].max()[0]}"
  )


  ax.annotate(texto_anotacion1, xy=(-0.1,45), xytext=(-0.4,50), fontsize=12, color=AZUL_1,
              bbox=dict(boxstyle="round,pad=0.3", edgecolor=GRIS_3, facecolor=BLANCO),
              arrowprops=dict(facecolor=AZUL_1, arrowstyle="->", connectionstyle="arc3,rad=0.3"))

  ax.annotate(texto_anotacion2, xy=(0.9,45), xytext=(0.6,50), fontsize=12, color=AZUL_1,
              bbox=dict(boxstyle="round,pad=0.3", edgecolor=GRIS_3, facecolor=BLANCO),
              arrowprops=dict(facecolor=AZUL_1, arrowstyle="->", connectionstyle="arc3,rad=0.3"))

  return plt
````
